# Load comments

Tried to do this with regex however SparkSQL does not support Unicode scalar values in its regular expression patterns. (At least according to copilot)

```python
%%sql
CREATE OR REPLACE TABLE Enhanced.videos AS
  SELECT *
  FROM 
  (
    SELECT
      id,
      title,
      playlistId,
      viewCount,
      commentCount,
      CASE
            WHEN title LIKE '%Albania%' THEN 'Albania'
            WHEN title LIKE '%Andorra%' THEN 'Andorra'
            WHEN title LIKE '%Armenia%' THEN 'Armenia'
            WHEN title LIKE '%Australia%' THEN 'Australia'
            WHEN title LIKE '%Austria%' THEN 'Austria'
            WHEN title LIKE '%Azerbaijan%' THEN 'Azerbaijan'
            WHEN title LIKE '%Belarus%' THEN 'Belarus'
            WHEN title LIKE '%Belgium%' THEN 'Belgium'
            WHEN title LIKE '%Bosnia and Herzegovina%' THEN 'Bosnia and Herzegovina'
            WHEN title LIKE '%Bulgaria%' THEN 'Bulgaria'
            WHEN title LIKE '%Croatia%' THEN 'Croatia'
            WHEN title LIKE '%Cyprus%' THEN 'Cyprus'
            WHEN title LIKE '%Czech Republic%' THEN 'Czechia'
            WHEN title LIKE '%Czechia%' THEN 'Czechia'
            WHEN title LIKE '%Denmark%' THEN 'Denmark'
            WHEN title LIKE '%Estonia%' THEN 'Estonia'
            WHEN title LIKE '%Finland%' THEN 'Finland'
            WHEN title LIKE '%France%' THEN 'France'
            WHEN title LIKE '%Georgia%' THEN 'Georgia'
            WHEN title LIKE '%Germany%' THEN 'Germany'
            WHEN title LIKE '%Greece%' THEN 'Greece'
            WHEN title LIKE '%Hungary%' THEN 'Hungary'
            WHEN title LIKE '%Iceland%' THEN 'Iceland'
            WHEN title LIKE '%Ireland%' THEN 'Ireland'
            WHEN title LIKE '%Israel%' THEN 'Israel'
            WHEN title LIKE '%Italy%' THEN 'Italy'
            WHEN title LIKE '%Latvia%' THEN 'Latvia'
            WHEN title LIKE '%Lithuania%' THEN 'Lithuania'
            WHEN title LIKE '%Luxembourg%' THEN 'Luxembourg'
            WHEN title LIKE '%Malta%' THEN 'Malta'
            WHEN title LIKE '%Moldova%' THEN 'Moldova'
            WHEN title LIKE '%Monaco%' THEN 'Monaco'
            WHEN title LIKE '%Montenegro%' THEN 'Montenegro'
            WHEN title LIKE '%Morocco%' THEN 'Morocco'
            WHEN title LIKE '%Netherlands%' THEN 'Netherlands'
            WHEN title LIKE '%North Macedonia%' THEN 'North Macedonia'
            WHEN title LIKE '%Norway%' THEN 'Norway'
            WHEN title LIKE '%Poland%' THEN 'Poland'
            WHEN title LIKE '%Portugal%' THEN 'Portugal'
            WHEN title LIKE '%Romania%' THEN 'Romania'
            WHEN title LIKE '%Russia%' THEN 'Russia'
            WHEN title LIKE '%San Marino%' THEN 'San Marino'
            WHEN title LIKE '%Serbia%' THEN 'Serbia'
            WHEN title LIKE '%Serbia and Montenegro%' THEN 'Serbia and Montenegro'
            WHEN title LIKE '%Slovakia%' THEN 'Slovakia'
            WHEN title LIKE '%Slovenia%' THEN 'Slovenia'
            WHEN title LIKE '%Spain%' THEN 'Spain'
            WHEN title LIKE '%Sweden%' THEN 'Sweden'
            WHEN title LIKE '%Switzerland%' THEN 'Switzerland'
            WHEN title LIKE '%Turkey%' THEN 'Turkey'
            WHEN title LIKE '%Ukraine%' THEN 'Ukraine'
            WHEN title LIKE '%United Kingdom%' THEN 'United Kingdom'
            WHEN title LIKE '%Yugoslavia%' THEN 'Yugoslavia'
            ELSE 'Unknown'
        END as Country
    FROM
      Raw.videos
  )
  WHERE Country <> 'Unknown';
```
